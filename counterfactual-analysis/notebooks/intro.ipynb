# Counterfactual Analysis

## Introduction

This project presents a benchmark of machine learning models on multiple
tabular datasets, with a focus on explainable artificial intelligence (XAI)
and counterfactual explanations.

The analysis is conducted on three commonly used datasets:
Breast Cancer Wisconsin, Adult Income, and Heart Disease.
For each dataset, several classifiers are evaluated, and the best-performing
model is further analyzed using explainability techniques.

## Methodology

### Datasets
Three tabular datasets are considered, each representing a binary
classification task. The datasets include numerical features and,
where applicable, categorical attributes.

### Models
The following classifiers are trained for each dataset:
- Logistic Regression
- Random Forest
- Gradient Boosting

Model performance is evaluated using accuracy on a held-out test set.
The best-performing model is selected for further analysis.

### Explainability Methods
Model decisions are analyzed using:
- Feature importance (for tree-based models)
- SHAP for global and local explanations
- LIME for local explanations

These methods provide complementary perspectives on model behavior.

### Counterfactual Explanations
Counterfactual explanations are generated using the DiCE framework.
For selected test instances, minimal changes to input features are identified
that lead to a different model prediction, enabling actionable and
human-understandable insights.
